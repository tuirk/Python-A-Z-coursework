## Introduction to Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.rcParams['figure.figsize'] = 8, 5 #applies to all figs
```

```python
#press tab
vis1 = sns.distplot(stats["UserCount"])
```

```python
vis1 = sns.distplot(stats["UserCount"], bins = 30)
```

```python
vis1 = sns.distplot(stats["UserCount"], bins = 5)
```

```python
vis2 = sns.boxplot(data = stats, x= "Income", y= "BirthRate")
```

```python
vis2 = sns.lmplot(data = stats, x= "UserCount", y= "BirthRate" )
```

```python
vis2 = sns.lmplot(data = stats, x= "UserCount", y= "BirthRate", fit_reg= False )
#press shift tab
```

```python
vis2 = sns.lmplot(data = stats, x= "UserCount", y= "BirthRate", \
                  hue = 'Income', fit_reg= False, size = 10)
```

```python
vis2 = sns.lmplot(data = stats, x= "UserCount", y= "BirthRate", \
                  hue = 'Income', fit_reg= False, size = 10, aspect = 2)
```

### Keyword Arguments

```python
vis3 = sns.lmplot(data = stats, x= "UserCount", y= "BirthRate", \
                  hue = 'Income', fit_reg= False, size = 6, \
                  scatter_kws = {"s":100} )
```
